# Внутрішнє добуток і його властивості

image.png

Внутрішній добуток це операція, яка асоціює два вектора з числом. Він є
узагальненнм поняття скалярного добутку і має властивості симетричності та лінійності.

Обчислимо внутрішній добуток за запропонованою формулою

$ <x, y> := x^ty= \sum_{i=1}^n x_iy_i = 1*3 + 2 * 4 = 11$

Внутрішній добуток має бути симетричним та лінійним.

Перевіримо що результат дотримується властивостей симетричності та лінійності.

Симетричність: $ <x, y> =  <y, x> $

Лінійність: $<x+z,y> = <x,y> +<z,y>$

А також що він є скалярним (dot product)

```python
import numpy as np

# Вектори
x = np.array([1, 2])
y = np.array([3, 4])
z = np.array([2, 1])

# Обчислимо внутрішній добуток
def inner_product(u,v):
  return u[0]*v[0] + u[1]*v[1]
print(f"Внутрішній добуток: {inner_product(x,y)}\n")

# Визначення скалярного добутку
def dot_product(u, v):
    return np.dot(u, v)
print(f"Скалярний добуток: {dot_product(x,y)}\n")

# 1. Перевірка лінійності
left = inner_product(x + z, y)
right = inner_product(x, y) + inner_product(z, y)

print("Адитивність перевірка:")
print(f"Left: {left}, Right: {right}")

# 2. Перевірка симетричності
xy = inner_product(x, y)
yx = inner_product(y, x)

print("\nСиметричність перевірка:")
print(f"x · y: {xy}, y · x: {yx}")
```

# Обчислення відстаней

image.png

$||x-y||_2 = \sqrt{(x_1-y_1)^2+(x_2-y_2)^2} = \sqrt(9+16) = 5 $

Ця формула відповідає довжині гіпотенузи прямокутного трикутника де $(x_1-y_1)$ і $(x_2-y_2)$ є довжинами катетів

```python
import numpy as np

# Вектори
x = np.array([1, 1])
y = np.array([4, 5])

# Визначення L2-норми (евклідової відстані)
def l2_distance(a, b):
    return np.linalg.norm(a - b)

# Обчислення відстані
distance = l2_distance(x, y)

print(f"L2-відстань між векторами x і y: {distance}")
```

Демонстрація інтуітивного розуміння довжини відрізка між двома точками

```python
import numpy as np
import matplotlib.pyplot as plt

# Координати точок
x = np.array([1, 1])
y = np.array([4, 5])

# Обчислення L2-відстані
distance = np.linalg.norm(x - y)

# Створення графіку
plt.figure(figsize=(8, 8))
plt.scatter(x[0], x[1], color='blue', label='Point x (1, 1)')
plt.scatter(y[0], y[1], color='red', label='Point y (4, 5)')

# Відрізок між точками
plt.plot([x[0], y[0]], [x[1], y[1]], color='green', linestyle='--', label=f'L2 Distance = {distance:.2f}')

# Налаштування осей
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xlim(0, 6)
plt.ylim(0, 6)

# Додавання підписів та легенди
plt.title("L2 Distance between Two Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

# 4. Визначення кута між векторами

image.png

```python
x = np.vstack([1,0])
y = np.vstack([0,1])

cosAngle = lambda x, y: (np.dot(x.T,y)/np.sqrt(np.dot(np.dot(x.T,x),np.dot(y.T,y))))[0][0]
angle = lambda cos: np.arccos(cos)*(180/np.pi)
angle(cosAngle(x,y))
```

# 6. Ортонормований базис і його застусування

image.png

Для того щоб перевірити що базис є ортонормованим, потрібно визначити чи є вектори ортогональними і нормалізованими.

**Перевірка ортогональності:** їх скалярний добуток дорівнює 0,

**Перевірка нормалізованості:**  довжина векторів дорівнює 1

```python
vectors = np.array([
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
])

def is_normalized(vector):
    return np.isclose(np.linalg.norm(vector), 1)

# Перевірка ортогональності
def is_orthogonal(vectors):
    n = len(vectors)
    for i in range(n):
        for j in range(i + 1, n):
            #скалярний добуток - 0
            if not np.isclose(np.dot(vectors[i], vectors[j]), 0):
                return False
    return True


# Перевірка ортонормальності
are_normalized = all(is_normalized(v) for v in vectors)
are_orthogonal = is_orthogonal(vectors)

if are_normalized and are_orthogonal:
    print("Вектори є ортонормованими.")
else:
    print("Вектори НЕ є ортонормованими.")
    if not are_orthogonal:
        print("Вектори не є ортогональними.")
    if not are_normalized:
        print("Вектори не нормалізовані.")
```

PCA - Principal component analysis - алгоритм для зменшення лінійної розмірності.
Алгоритм використовується для стиснення даних та візуалізації, а також для
ідентифікації простих закономірностей, прихованих факторів та структури багатовимірних даних.


Візьмемо випадок коли нам потрібно стиснути дані. Нехай умовно уявимо що це точки у певному просторі, які мають певні координати, вони розкидані по нашому просторі у хаотичному порядку.
Для того щоб зменшити кількість координат - ми можемо спроектувати їх на **ортонормований базис.**

Отже наша ціль - знайти такий **ортонормований базис**, щоб спроектувати якомога більше таких точок.

Точки, які не попадають на цей **ортонормований базис** - відкидаються, і за рахунок цього відбувається стиснення.
